```python
import tensorflow as tf
```

Computational Graph
-------------
### (1) Build graph using TF operation

```python
# X and Y data
x_train = [1, 2, 3]
y_train = [1, 2, 3]

W = tf.Variable(tf.random_normal([1]), name='weight')
# 텐서플로에서 학습해서 변경하는 변수 = trainable variable
# W는 변수이므로 random하게 시작, shape은 일단 1
b = tf.Variable(tf.random_normal([1]), name='bias')

# Our hypothesis XW + b
hypothesis = x_train * W + b
```

```python
# cost/Loss function
cost = tf.reduce_mean(tf.square(hypothesis - y_train))
# reduce_mean, square 함수 이용하여 cost function 설정
# reduce_mean 함수는 합을 평균 내준다
```

```python
# GradientDescent
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
# magic이라고 생각하면 됨
# 텐서플로에서 제공하는 GradientDescentOptimizer 방법 이용할 것

train = optimizer.minimize(cost)
# optimizer에서 cost를 minimize하라
```

### (2~3) Run/update graph and get results

```python
# Launch the graph in a session
sess = tf.Session()
# Initializes global variables in the graph
sess.run(tf.global_variables_initializer())
# W와 b라는 variable을 정의해서 사용하기 전에는 반드시 initialize

# Fit the line
for step in range(2001):
    sess.run(train)
    if step % 20 == 0:
        print(step, sess.run(cost), sess.run(W), sess.run(b))
# 2000번 정도 돌리는데 20번에 한번씩 cost, W, b 프린트하기
# 학습이 진행될수록 cost 감소 & W는 1에 수렴 & b는 0으로
```

Placeholder (복습)
-------------

```python
a = tf.placeholder(tf.float32)
b = tf.placeholder(tf.float32)
adder_node = a + b
# + provides a shortcut for tf.add(a, b)

print(sess.run(adder_node, feed_dict={a: 3, b: 4.5}))
print(sess.run(adder_node, feed_dict={a: [1,3], b: [2,4]}))

# placeholder로 node를 만들어줄 경우 feed_dict에 값을 넘겨준다
```

Build graph with placeholders
-------------

```python
W = tf.Variable(tf.random_normal([1]), name='weight')
b = tf.Variable(tf.random_normal([1]), name='bias')
X = tf.placeholder(tf.float32)
Y = tf.placeholder(tf.float32)

hypothesis = X * W + b
cost = tf.reduce_mean(tf.square(hypothesis - Y))
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
train = optimizer.minimize(cost)

sess = tf.Session()
sess.run(tf.global_variables_initializer())

for step in range(2001):
    cost_val, W_val, b_val, _ = sess.run([cost, W, b, train],
                                         feed_dict={X: [1, 2, 3], Y: [1, 2, 3]})
    if step % 20 ==0:
        print(step, cost_val, W_val, b_val)
        
# cost, W, b를 각각 _value로 저장한다
# train 값은 필요 없으니 _로 처리한다
```

Full code
-------------

```python
import tensorflow as tf
W = tf.Variable(tf.random_normal([1]), name='weight')
b = tf.Variable(tf.random_normal([1]), name='bias')
X = tf.placeholder(tf.float32, shape=[None])
Y = tf.placeholder(tf.float32, shape=[None])

hypothesis = X * W + b
cost = tf.reduce_mean(tf.square(hypothesis - Y))
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
train = optimizer.minimize(cost)

sess = tf.Session()
sess.run(tf.global_variables_initializer())

for step in range(2001):
    cost_val, W_val, b_val, _ = sess.run([cost, W, b, train],
                                        feed_dict={X: [1,2,3,4,5], Y:[2.1,3.1,4.1,5.1,6.1]})
    if step % 20 ==0:
        print(step, cost_val, W_val, b_val)
```

```python
# Testing our model
print(sess.run(hypothesis, feed_dict={X: [5]}))
print(sess.run(hypothesis, feed_dict={X: [2.5]}))
print(sess.run(hypothesis, feed_dict={X: [1.5,3.5]}))
```
